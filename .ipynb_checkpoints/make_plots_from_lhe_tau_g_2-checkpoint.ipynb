# MadGraph Project

MadGraph was used to simulate 100 eetautau final state events, from a e plus e minus collider. 

----------------
Syntax for generating the events in Madgraph:

output ee_eetautau

launch ee_eetautau

```python
import pylhe
import math
from pprint import *
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from uproot_methods import TLorentzVector
matplotlib.rcParams['figure.dpi']= 150
```

```python
def calc_pt(px,py):
    '''Calculates transverse momentum'''
    return math.sqrt(px**2+py**2)

def my_hist_plotter(data,x_min,x_max,
                    n_bins,x_axis_name,save,save_name):
    '''Makes basic histogram of data with range specified'''
    fig= plt.figure()
    plt.hist(data,bins=n_bins,range=[x_min, x_max],
             histtype='step',color='tab:blue',fill = True,
             linewidth=0,alpha=0.2, normed = False)
    plt.xlabel(x_axis_name)
    plt.ylabel("Events")
    if save:
        plt.savefig(save_name+".pdf",transparent=True)
    
def my_hist_plotter2(data,n_bins,x_axis_name,save=False,save_name=None,my_range = None):
    '''Makes basic histogram of data without range specified'''
    fig= plt.figure()
    plt.hist(data,bins=n_bins,histtype='step',range=my_range,
             color='tab:blue',fill = True,
             linewidth=0,alpha=0.2)
    plt.xlabel(x_axis_name)
    plt.ylabel("Events")

    if save:
        plt.savefig(save_name+".pdf",transparent=True)

name='unweighted_events.lhe'
```

```python
# What sort of things to look at? 
# invariant mass of e+e- pairs
# missing momentum
```

## PDG particle IDs 
- Ap = 2001
- e_minus = 11
- e_plus = -11

## PDG particle status
- Status = 1 means final state particle



```python
invariant_mass_0  = []
invariant_mass_1 = []

# Basic event loop 
for event in pylhe.readLHE(name):
    # Initialize electron and positrons
    ep0 = None
    ep1 = None
    em0 = None
    for particle in event.particles:
        print(particle.id)
        if particle.status == 1.0: # final state partilce 
            px = particle.px
            py = particle.py
            pz = particle.pz
            m = particle.m
            p = TLorentzVector.from_xyzm( px, py, pz, m)            
            if particle.id == -11: # positron
                if ep0 == None: 
                    ep0 = p
                else:
                    ep1 = p
            if particle.id == 11: # electron 
                em0 = p
    #print(ep0, ep1, em0)
    pairing_0 = ep0+em0
    pairing_1 = ep1+em0
    invariant_mass_0 += [pairing_0.mass*1000.0] # pairing 0 
    invariant_mass_1 += [pairing_1.mass*1000.0] # pairing 1 (MeV)
               
```

```python
# 2D plots 
from matplotlib.colors import LogNorm
def make_2d_hist(x, y, x_label="", y_label="",):
    plt.figure()
    plt.hist2d(x,y,
               range=[[3,5],[0,5]],
               bins=10, 
               norm=LogNorm(),
               cmap='viridis')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.colorbar(label="Events")
    #plt.title("Linear fit on training set")
    plt.legend(loc="upper left",frameon=False)
    return
```
